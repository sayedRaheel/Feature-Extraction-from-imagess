### Create a train data loader that returns image arrays and labels

```python
#Creating a train data loader that returns image arrays and labels
def load_data2(data_dir):
    images = []
    labels = []
    
    # Define a dictionary to map label names to their corresponding numeric values
    label_dict = {'Score_1': 0, 'Score_2': 1, 'Score_3': 2,'Score_4':3}
    
    # Iterate over the directories in the data directory
    for dir_name in os.listdir(data_dir):
        dir_path = os.path.join(data_dir, dir_name)
        
        # Skip any non-directory files
        if not os.path.isdir(dir_path):
            continue
        
        # Get the label for this directory based on its name
        label = label_dict.get(dir_name, None)
        if label is None:
            continue
        
        # Iterate over the files in the directory
        for file_name in os.listdir(dir_path):
            file_path = os.path.join(dir_path, file_name)
            
            # Skip any non-image files
            if not file_path.endswith('.jpg'):
                continue
            
            # Read in the image file and append it to the images list along with its label
            image = cv2.imread(file_path)
            image = cv2.resize(image, (224,224))
            images.append(image)
            labels.append(label)
            
    return print('Image array dimensions of Train Dataset:', np.shape(images)), print('Label dimensions:', np.shape(labels))
```

### Print image arrays, labels and file names dimensions

```python
import os
import cv2

def Test_load_data(data_dir):
    images = []
    labels = []
    file_names = []

    # Iterate over the directories in the data directory
    for dir_name in os.listdir(data_dir):
        dir_path = os.path.join(data_dir, dir_name)

        # Skip any non-directory files
        if not os.path.isdir(dir_path):
            continue

        # Iterate over the files in the directory
        for file_name in os.listdir(dir_path):
            file_path = os.path.join(dir_path, file_name)
            file_names.append(file_name)

            # Skip any non-image files
            if not file_path.endswith('.jpg'):
                continue

            # Read in the image file and append it to the images list
            image=cv2.imread(file_path)
            image = cv2.resize(image, (224,224))
            images.append(image)

            # Append the label to the labels list
            labels.append(dir_name)

    # Combine the images, labels, and file names into a list of tuples
    data = list(zip(images, labels, file_names))

    # Print the dimensions of the images, labels, and file names
    print("DIMENSION:", np.shape(images))
    print("Dimension of combined images with their labels", np.shape(data))
    print("File_Name:", file_names[:1])

    # Return the data
    return data
```

### Feature extraction for training Data

```python
# Extracting features of training  images using HOG
import os
import cv2
import numpy as np
from skimage.feature import hog

def load_data_hog(data_dir):
    images_for_hog = []
    labels = []
    file_names = []
    for dir_name in os.listdir(data_dir):
        dir_path = os.path.join(data_dir, dir_name)
        if not os.path.isdir(dir_path):
            continue
        for file_name in os.listdir(dir_path):
            file_path = os.path.join(dir_path, file_name)
            if not file_path.endswith('.jpg'):
                continue
            img = cv2.imread(file_path)
            img = cv2.resize(img, (512,512))
            images_for_hog.append(img)
            labels.append(dir_name)
            file_names.append(file_name)
    return images_for_hog

def extract_features(images):
    features = []
    for image in images:
        hog_features = hog(image, orientations=9, cells_per_block=(2, 2), multichannel=True)
        features.append(hog_features)
    return features

```

```python

import cv2

# Function to extract SIFT features from training images
def extract_sift_train(train_images):
    # Create SIFT object
    sift = cv2.xfeatures2d.SIFT_create()
    # Create an empty list to hold the descriptors of each image
    descriptors_list = []

    # Loop over each image in the training set
    for image in train_images:
        # Detect keypoints and compute SIFT descriptors for the image
        kp, des = sift.detectAndCompute(image, None)
        # If the descriptors are not empty, add them to the list
        if des is not None:
            descriptors_list.append(des)

    # Concatenate the descriptors of all images in the list into a single numpy array
    descriptors_train = np.concatenate(descriptors_list, axis=0)

    # Print the size of the training features and the total number of features
    print("Training features:",len(train_images),'*', np.shape(descriptors_train))

    # Return the training descriptors
    return descriptors_train

# Function to extract SIFT features from test images
def extract_sift_test(test_images):
    # Create SIFT object
    sift = cv2.xfeatures2d.SIFT_create()
    # Create an empty list to hold the descriptors of each image
    descriptors_list = []

    # Loop over each image in the test set
    for image in test_images:
        # Detect keypoints and compute SIFT descriptors for the image
        kp, des = sift.detectAndCompute(image, None)
        # If the descriptors are not empty, add them to the list
        if des is not None:
            descriptors_list.append(des)

    # Concatenate the descriptors of all images in the list into a single numpy array
    descriptors_test = np.concatenate(descriptors_list, axis=0)
    
    # Print the size of the test features and the total number of features
    print("Test features:",len(test_images),'*',np.shape(descriptors_test))


    # Return the test descriptors
    return descriptors_test
```

```python
import cv2

# Function to extract SURF features from training images
def extract_surf_train(train_images):
    # Create SURF object
    surf = cv2.xfeatures2d.SURF_create(extended=True)
    # Create an empty list to hold the descriptors of each image
    descriptors_list = []

    # Loop over each image in the training set
    for image in train_images:
        # Detect keypoints and compute SURF descriptors for the image
        kp, des = surf.detectAndCompute(image, None)
        # If the descriptors are not empty, add them to the list
        if des is not None:
            descriptors_list.append(des)

    # Concatenate the descriptors of all images in the list into a single numpy array
    descriptors_train = np.concatenate(descriptors_list, axis=0)

    # Print the size of the training features and the total number of features
    print("Training features:",len(train_images),'*', np.shape(descriptors_train))

    # Return the training descriptors
    return descriptors_train

# Function to extract SURF features from test images
def extract_surf_test(test_images):
    # Create SURF object
    surf = cv2.xfeatures2d.SURF_create(extended=True)
    # Create an empty list to hold the descriptors of each image
    descriptors_list = []

    # Loop over each image in the test set
    for image in test_images:
        # Detect keypoints and compute SURF descriptors for the image
        kp, des = surf.detectAndCompute(image, None)
        # If the descriptors are not empty, add them to the list
        if des is not None:
            descriptors_list.append(des)

    # Concatenate the descriptors of all images in the list into a single numpy array
    descriptors_test = np.concatenate(descriptors_list, axis=0)


    # Print the size of the test features and the total number of features
    print("Test features:",len(test_images),'*',np.shape(descriptors_test))


    # Return the test descriptors
    return descriptors_test
```
